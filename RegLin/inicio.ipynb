```python
import numpy as np  
import matplotlib.pyplot as plt
```

```python
x = np.array([0,1,2,3])
y = np.array([0,1,2,3])
```

```python
def predict(alpha,beta,x):
    return alpha*x + beta
```

```python
alpha = 1
beta=0

ÿ = predict(alpha=alpha,beta=beta,x=x)


fig,ax = plt.subplots(figsize=(5,5))
ax.scatter(x,y,c="b")
plt.plot(x,ÿ,c="r")
```
